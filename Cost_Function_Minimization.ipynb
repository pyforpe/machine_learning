## Cost Function Minimization for Linear Equation

Initially Take X and Y

```python
x = [i for  i in range(10)]
y = [i*3+2 for  i in range(10)]
print(x)
print(y)
```

Assume theta0 and theta1 for initialization

```python
theta0 = [1]
theta1 = [1]
```

Write function that can give predicted value of y from given x using 
      y = theta0 + theta1 * x

```python
def pred_y(theta0, theta1, x):
    lst = []
    for i in x:
        lst.append(theta0 + theta1*i)
    return lst
```

Define cost function

```python
def cost_func(pred_y, y):
    lst = []
    for i in range(len(y)):
        lst.append((pred_y[i] - y[i])**2)
    total_cost = 0
    for cost in lst:
        total_cost += cost
    return total_cost/(2*len(y))
```

```python
def theta0_params(pred_y, y):
    total = 0
    for i in range(len(y)):
        total += (pred_y[i] - y[i])
    return total
```

```python
def theta1_params(pred_y, y, x):
    total = 0
    for i in range(len(y)):
        total += (pred_y[i] - y[i]) * x[i]
    return total
```

Alpha is called learning rate and assumed 0.01

```python
alpha = 0.01
cost = []
for i in range(10000):
    theta0.append(theta0[-1] - alpha * theta0_params(pred_y(theta0[-1], theta1[-1], x), y) / len(y))
    theta1.append(theta1[-1] - alpha * theta1_params(pred_y(theta0[-2], theta1[-1], x), y, x) / len(y))
    cost.append(cost_func(pred_y(theta0[-1], theta1[-1], x), y))
```

```python
print(pred_y(theta0[-1], theta1[-1], x))
print(y)
```
